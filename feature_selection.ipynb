```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import f_regression
```

```python
x_train = pd.read_csv('csv/x_train.csv')
y_train = pd.read_csv('csv/y_train.csv', header=None)
x_test = pd.read_csv('csv/x_test.csv')
y_test = pd.read_csv('csv/y_test.csv', header=None)
```

```python
f, pval = f_regression(x_train, np.ravel(y_train), center=True)
```

```python
p = .01
x_train_new = x_train[cols[pval<p]]
x_test_new = x_test[cols[pval<p]]
```

```python
x_train_new.to_csv('csv/x_train_significant.csv', index=None)
x_test_new.to_csv('csv/x_test_significant.csv', index=None)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# Old acc
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
result = list(y_pred) == y_test
(sum(result[0])+0.0)/len(result)
```

```python
# New acc
model.fit(x_train_new, y_train)
y_pred = model.predict(x_test_new)
result = list(y_pred) == y_test
(sum(result[0])+0.0)/len(result)
```

```python
# New acc
model.fit(x_train_new, y_train)
y_pred_probs = model.predict_proba(x_test_new)
y_pred_probs = [x[1] for x in y_pred_probs]
y_pred_probs
```
